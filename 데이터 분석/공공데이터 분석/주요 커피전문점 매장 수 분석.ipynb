```python
import pandas as pd
from glob import glob
```

```python
file_names = glob("C:/Users/yj/Desktop/data/*.csv")  # *.csv : csv로 끝나는 모든 파일
file_names
```

```python
total = pd.DataFrame()

for file_name in file_names:
    temp = pd.read_csv(file_name, sep=',', encoding='utf-8')
    total = pd.concat([total,temp])
```

```python
# concat을 하면 인덱스도 모두 붙어 버림. 따라서 나중에 인덱스를 사용하여 인덱싱할 수 없으니 인덱스를 리셋 시켜야 함.
total.reset_index(inplace=True, drop=True)  # drop = True 기존 인덱스 없애기
total.head(2)
```

```python
data = total[['상호명', '지점명', '상권업종대분류명', '상권업종중분류명', '시도명',
             '시군구명', '행정동명']]
data.head(2)
```
