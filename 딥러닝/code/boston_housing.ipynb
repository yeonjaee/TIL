```python
from tensorflow.keras.datasets.boston_housing import load_data

(x_train, y_train), (x_test,y_test) = load_data(path='boston_housing.npz', test_split=0.2,seed=777)
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
import numpy as np

# 데이터 표준화
mean = np.mean(x_train, axis=0)
std = np.std(x_train, axis=0)

x_train = (x_train - mean) / std
x_test = (x_test - mean) / std

# 검증 데이터 셋 생성
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.33, random_state = 777)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

# 13차원의 데이터를 입력받고, 64개의 출력을 가지는 첫 번째 Dense
model.add(Dense(64,activation='relu', input_shape=(13, )))

model.add(Dense(32,activation='relu'))  # 32개의 출력을 가지는 Dense ㅡㅇ

model.add(Dense(1)) # 하나의 값을 출력

model.compile(optimizer = 'adam', loss = 'mse', metrics=['mae'])
```

```python
history = model.fit(x_train, y_train, epochs=300, validation_data=(x_val,y_val))
model.evaluate(x_test, y_test)
```

```python
from tensorflow.keras.datasets.boston_housing import load_data
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
# K-폴드 사용
from sklearn.model_selection import KFold

(x_train, y_train), (x_test,y_test) = load_data(path='boston_housing.npz', test_split=0.2,seed=777)

# 데이터 표준화
mean = np.mean(x_train, axis=0)
std = np.std(x_train, axis=0)

x_train = (x_train - mean) / std
x_test = (x_test - mean) / std

# K-폴드 진행
k = 3

# 주어진 데이터셋을 k 만큰 등분
# 여기서는 3이므로 학습 데이터셋(404개)을 3등분하여
# 한 개는 검증셋으로, 나머지 두 개는 학습 데이터셋으로 활용한다.
kfold = KFold(n_splits=k)

# K-폴드 과정에서 재사용을 위해 모델을 반환하는 함수를 정의한다.
def get_model():
  model = Sequential()
  model.add(Dense(64,activation='relu', input_shape=(13, )))
  model.add(Dense(32,activation='relu'))
  model.add(Dense(1))
  model.compile(optimizer = 'adam', loss = 'mse', metrics=['mae'])
  return model

mae_list = [] # 테스트 셋을 평가한 후 결과를 저장

# k번 진행
for train_index, val_index in kfold.split(x_train):
  # 해당 인덱스는 무작위로 생성
  # 과대적합을 피하기 위함
  x_train_fold, x_val_fold = x_train[train_index], x_train[val_index]
  y_train_fold, y_val_fold = y_train[train_index], y_train[val_index]

  # 모델 불러오기
  model = get_model()

  model.fit(x_train_fold, y_train_fold, epochs = 300, validation_data = (x_val_fold, y_val_fold))
  _, test_mae = model.evaluate(x_test,y_test)
  mae_list.append(test_mae)
```

```python
print(mae_list)
```

```python
print(np.mean(mae_list))
```
