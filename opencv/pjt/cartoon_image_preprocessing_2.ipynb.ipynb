```python
import os
import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

## 원본 이미지

```python
dog_img = cv2.imread('C:/Users/yj/Desktop/2_3.jpg')
dog_img = cv2.cvtColor(dog_img,cv2.COLOR_BGR2RGB)
plt.imshow(dog_img)
dog_img.shape
```

## resize
모든 트레이닝 이미지를 256x256 사이즈로 resizing

```python
path = 'C:/Users/yj/Desktop/mydog/'

img_size = 256

for img in tqdm(os.listdir(path)):
    img_name = img
    img = cv2.imread(path+img)
    img = cv2.resize(img,(img_size,img_size))
    cv2.imwrite(f'C:/Users/yj/Desktop/mydog_resized/{img_name}',img)
```

```python
dog_img = cv2.imread('C:/Users/yj/Desktop/mydog_resized/104.jpg')
dog_img = cv2.cvtColor(dog_img,cv2.COLOR_BGR2RGB)
plt.imshow(dog_img)
dog_img.shape
```

## GaussianBlur 처리

```python
path = 'C:/Users/yj/Desktop/ponyo/'

for img in tqdm(os.listdir(path)):
    img_name_blur = img
    img = cv2.imread(path+img)
    img = cv2.GaussianBlur(img, (0, 0), 3)
    cv2.imwrite(f'C:/Users/yj/Desktop/trainB_smooth/{img_name_blur}',img)
```

```python
dog_img = cv2.imread('C:/Users/yj/Desktop/mydog_blurred/104.jpg_blur')
dog_img = cv2.cvtColor(dog_img,cv2.COLOR_BGR2RGB)
plt.imshow(dog_img)
dog_img.shape
```
