## OpenCV 주요함수 설명

#### Opencv tutorials: https://docs.opencv.org/master/

cv2.imread(filename. flags = None) -> retval

flags:
- cv2.IMREAD_COLOR: BGR color로 읽기

- cv2.IMREAD_GRAYSCALE: 그레이 color로 읽기

- cv2.IMREAD_UNCHANGED: 파일 속성대로 읽기

- cv2.IMREAD_REDUCED_GRAYSCALE_2: 그레이 제거하고 1/2 크기로 읽기

- cv2.IMREAD_REDUCED_COLOR_2 

retval: numpy.ndarray로 반환

```python
img = cv2.imread('./fig/puppy_1280_853.jpg', cv2.IMREAD_REDUCED_GRAYSCALE_8)

if img is None:
    print('Image read failed')
    sys.exit()
    
cv2.namedWindow('puppy')
cv2.imshow('puppy', img)

# Keyboard 입력 (모든 키) 을 기다리는 동안 영상을 출력 
cv2.waitKey()

# 키보드 입력을 받으면 화면에 나타난 모든 창을 닫음
cv2.destroyAllWindows()
```

### image 저장하기: cv2.imwrite()

```python
# cv2.imwrite(filename, img, params =None) -> retval

# filename: 저장할 파일 이름
# img: 저장할 자료 (numpy.ndarray)
# params: 파일저장 옵션
# retval: 성공 True, 실패 False

img  = cv2.resize(img, (640, 480), interpolation = cv2.INTER_AREA)

cv2.imwrite('fig/puppy.png', img)
```

### 새창띄우기 : cv2.namedWindow()

```python
# cv2.namedWindow(winname, flags = None) -> None

# winname: 창 이름
# flags:
# cv2.WINDOW_NORMAL: 영상크기를 창 크기에 맞게 지정
# cv2.WINDOW_AUTOSIZE: 창크기를 영상 크기에 맞게 변경

# cv2.namedWindow('puppy', cv2.WINDOW_NORMAL)
cv2.namedWindow('puppy', cv2.WINDOW_AUTOSIZE)
cv2.imshow('puppy', img)

cv2.waitKey()
cv2.destroyAllWindows()
# cv2.destroyWindow('winname')

# 창의 위치 변경
# cv2.moveWindow('winname', x, y) -> None

# 창의 사이즈 변경
# namedWindow('winname', flags = cv2.WINDOW_NORMAL)
# cv2.resizeWindow('winname', width, height) -> None
```

### matplotlib 라이브러리를 이용한 영상출력

```python
# img = cv2.imread('fig/cat.bmp')
imgBGR = cv2.imread('fig/puppy.bmp', cv2.IMREAD_COLOR)


if imgBGR is None:
    print('Image read failed')
    sys.exit()


# BGR to RGB
imgRGB = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)

## plt를 이용한 영상출력
# plt.imshow(imgBGR)
plt.imshow(imgRGB)
plt.axis('off')
plt.show()
```

## 이미지 슬라이드쇼

```python
import sys
import glob
import cv2
import os

#list 형태로 반환
img_list_os = os.listdir('fig/images')

img_files = []
for i in img_list_os:
    img_path = './fig/images/' + i
    img_files.append(img_path)
    

# img_files_glob = glob.glob('fig/images/*.jpg')
# print(img_files_glob)

for i in img_files:
    print(i)


# # 전체 화면으로 'image' 창 생성
cv2.namedWindow('scene', cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN,
                    cv2.WINDOW_FULLSCREEN)

for i in img_files:
    img = cv2.imread(i)
    
    cv2.imshow('scene', img)
    
    if cv2.waitKey(1000) == 27:
        break
    
cv2.destroyAllWindows()
```

```python
import sys
## Wild card
import glob
import cv2

img_files = glob.glob('fig/images/*.jpg')

for f in img_files:
    print(f)
    
# 전체 화면으로 'image' 창 생성
cv2.namedWindow('scene',cv2.WINDOW_NORMAL)
cv2.setWindowProperty('scene', cv2.WND_PROP_FULLSCREEN, cv2.WINDOW_FULLSCREEN)

# for i in img_files:
#     img = cv2.imread(i)
    
#     cv2.imshow('scene', img)
    
#     if cv2.waitKey(1000) == 27:
#         break

cnt = len(img_files)
idx = 0

while True:
    img = cv2.imread(img_files[idx])
#     img = cv2.resize(img, dsize = (1920, 1080), interpolation = cv2.INTER_AREA)
    
    cv2.imshow('image', img)
    
    if cv2.waitKey(1000) == 27: #ESC
        break
    
    idx += 1
    if idx >= cnt:
        idx = 0
        
cv2.destroyAllWindows()
```
